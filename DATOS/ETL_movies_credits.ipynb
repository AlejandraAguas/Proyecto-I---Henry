```python
#Importar las Librerías. 
#Se importaron las siguientes librerías.
import pandas as pd
import re 
import json
import ast
```

```python
#Cargar el Conjunto de Datos 
df = pd.read_csv("movies_dataset.csv")
df
```

```python
#Desanidar los campos “id” y “name” de la columna “belongs_to_collection” 
#y se crearon nuevas columnas “id_coleccion” y “nombre_coleccion”. Lo mismo se hizo para la columna “production_companies”.

busq_1 = r"'id': ([^,]*)"
df['id_coleccion'] = df['belongs_to_collection'].str.extract(busq_1, flags=re.IGNORECASE)
busq_2 = r"'name': '([^']*)"
df['nombre_coleccion'] = df['belongs_to_collection'].str.extract(busq_2, flags=re.IGNORECASE)
df
```

```python

busq_3 = r"'id': ([^}]*)"
df['id_produccion'] = df['production_companies'].str.extract(busq_3, flags=re.IGNORECASE)
busq_4= r"'name': '([^']*)"
df['nombre_produccion'] = df['production_companies'].str.extract(busq_4, flags=re.IGNORECASE)
df
```

```python
#Verificar las columnas “revenue” y “budget” para identificar valores nulos. 
#Los valores nulos en “revenue” se reemplazaron por 0. 
#Además, se eliminan filas con valores nulos en la columna “release_date”.
nulos_revenue = df["revenue"].isnull().sum()
print(nulos_revenue)

nulos_budget = df["budget"].isnull().sum()
print(nulos_budget)
```

```python
#Reemplazar los valores nulos de la columna "revenue" por 0
df["revenue"] = df["revenue"].fillna(0)
df
```

```python
#Verificar los valores nulos de todas las columnas
df.isnull().sum().sort_values(ascending=False)
```

```python
#Eliminar los valores nulos de la columna "relase_date"
df = df.dropna(subset=["release_date"])
df
```

```python
#Verificar los valores nulos de todas las columnas
df.isnull().sum().sort_values(ascending=False)
```

```python
df['release_date'] = pd.to_datetime(df['release_date'], errors='coerce').dt.strftime('%Y-%m-%d')
df['release_date']
```

```python
#Extraer el año de estreno y lo asignamos a una nueva columna
df['release_year'] = pd.to_datetime(df['release_date'], errors='coerce').dt.year.astype('Int64')
df
```

```python
#Convertir las columnas "revenue" y "budget" a tipo numérico
df['revenue'] = pd.to_numeric(df['revenue'], errors= "coerce")
df['budget'] = pd.to_numeric(df['budget'], errors= "coerce")

#Se creo y se calculo en la columna de return el retorno de inversion
df['return'] = df.apply(lambda row: row['revenue'] / row['budget'] if row['budget'] != 0 else 0, axis=1)
df
```

```python
#Eliminar las columnas "video", "imdb_id", "adult", "original_title", "poster_path", "homepage", "belongs_to_collection", "production_companies"
eliminar_columnas = ["video", "imdb_id", "adult", "original_title", "poster_path", "homepage", "belongs_to_collection", "production_companies"]

# Eliminar columnas con drop
df2 = df.drop(columns=eliminar_columnas)
```

```python
#visualizar las columnas completas
pd.set_option('display.max_columns', None)
df2
```

```python
#Convertir la cadena en una lista de diccionarios
def extraer_nombres_genero(generos_str):
    try:
        generos_lista = json.loads(generos_str.replace("'", '"'))  # Reemplazar comillas simples por comillas dobles
        nombres_genero = [genero['name'] for genero in generos_lista]
        return ', '.join(nombres_genero)
    except (json.JSONDecodeError, KeyError):
        return None  # Manejar errores de formato o claves faltantes

#Aplicar la función a la columna 'genres'
df2['genres_name'] = df2['genres'].apply(extraer_nombres_genero)

print(df2['genres_name'])
```

```python
#Eliminar las columnas "genres"
eliminar_genres = ["genres"]
movies= df2.drop(columns=eliminar_genres)

print(movies.shape)
```

```python
#Verificar los "id" duplicados y los eliminamos
movies = movies.drop_duplicates(subset='id', keep='first')
movies
```

```python
#cargar el dataset de Credits
credits = pd.read_csv("credits.csv")
credits
```

```python
#Verificar los "id" duplicados y los eliminamos
credits = credits.drop_duplicates(subset='id', keep='first')
credits
```

```python
#convertir la columna cast de texto a una lista de diccionarios
credits['cast'] = credits['cast'].apply(ast.literal_eval)

#definir una función para extraer los nombres de los actores
def extract_actor_names(cast_list):
    return [member['name'] for member in cast_list if 'name' in member]

#aplicar la función a la columna cast
credits['actores'] = credits['cast'].apply(extract_actor_names)
```

```python
def safe_literal_eval(text):
    try:
        return ast.literal_eval(text)
    except (ValueError, SyntaxError):
        return []

#convertir la columna 'crew' de texto a una lista de diccionarios
credits['crew'] = credits['crew'].apply(safe_literal_eval)

#definir una función para extraer el nombre del director
def extract_director(crew_list):
    for member in crew_list:
        if isinstance(member, dict) and member.get('job') == 'Director':
            return member.get('name')
    return None

#aplicar la función a la columna 'crew'
credits['director'] = credits['crew'].apply(extract_director)

#Mostrar el DataFrame con la nueva columna
print(credits[['id', 'director']])
```

```python
#Eliminar las columnas "cast", "crew"
eliminar_columnas2 = ["cast", "crew"]
data_credits = credits.drop(columns=eliminar_columnas2)

print(data_credits.shape)
```

```python
#Convertir la columna 'id' a string en ambos DataFrames
movies['id'] = movies['id'].astype(str)
data_credits['id'] = data_credits['id'].astype(str)
```

```python
#unir los dos dataset
data_fusionado = movies.merge(data_credits, how='inner', on='id')
data_fusionado
```

```python
#Crear una función para verificar si el título contiene solo caracteres alfabéticos
def filtrar_titulo(titulo):
    #Verificar si el título contiene solo letras y espacios
    return re.match("^[a-zA-Z0-9\\s\\$\\-\\#\\?\\&\\!\\ñ\\¿\\¡]*$", titulo) is not None

#Filtrar las filas donde el título de la película es válido
df_fusionado = data_fusionado[data_fusionado['title'].apply(filtrar_titulo)]
```

```python
#Convertir todos los valores en la columna 'title' a cadenas
data_fusionado['title'] = data_fusionado['title'].astype(str)

# Filtrar las filas donde el título de la película es válido
df_fusionado = data_fusionado[data_fusionado['title'].apply(filtrar_titulo)]
```

```python
#Eliminar las columnas "production_countries", "spoken_languages",	"status", "tagline"
eliminar_col = ["production_countries", "spoken_languages",	"status", "tagline"]
df_combinado = df_fusionado.drop(columns=eliminar_col)
```
